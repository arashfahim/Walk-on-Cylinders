```python
import numpy as np
import random
import time
import os
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
path = os.getcwd()
dx = 0.01
M = 8
samples = {}
steps = [np.array([dx,0]),np.array([-dx,0]),np.array([0,dx]),np.array([0,-dx])]
for m in range(M):
    u = np.array([[0.0,0.5]])
    v = u-np.array([[0,0.5]])
    n = 0
    while (v**2).sum() < 4:
        if n == 0:
            t = u
        else:
            t = np.concatenate((t,u), axis = 0)
        n += 1
        u += random.choice(steps)
        v = u-np.array([[0,0.5]])
    samples[str(m+1)] = t
```

```python
plt.figure(figsize=(10,10),dpi=300)
for k in samples:
    t = samples[k]
    plt.plot(t[:,0],t[:,1]);
    plt.plot(t[-1,0],t[-1,1],marker='x',linewidth=8, markersize=8, color='red');
c = plt.Circle((0, 0.5), 2, color='pink', alpha=0.5, fill=True);
plt.plot([0],[0],marker='*', color='r',linewidth=8);
a =  plt.gca()
a.add_patch(c)
xax = a.axes.get_yaxis()
xax = xax.set_visible(False)
plt.axis('off')
plt.rc('axes.spines', **{'bottom':True, 'left':True, 'right':False, 'top':False})
# plt.savefig("bm_2d_1.png");
```

```python
path = os.getcwd()
dx = 0.02
M = 1000
N = 1000
samples = {}
steps = [np.array([dx,0]),np.array([-dx,0]),np.array([0,dx]),np.array([0,-dx])]
avg = []
for m in range(M):
    u = np.array([[0.0,1.8]])
    v = u-np.array([[0,0.5]])
    n = 0
    while (v**2).sum() < 4 and n <= N:
        t = u
        n += 1
        u += random.choice(steps)
        v = u-np.array([[0,0.5]])
        if n == M:
            avg.append(u)
    samples[str(m+1)] = t

plt.figure(figsize=(10,10),dpi=300)
for k in samples:
    t = samples[k]
    # plt.plot(t[:,0],t[:,1]);
    plt.plot(t[-1,0],t[-1,1],marker='+',linewidth=8, markersize=8, color='red');
c = plt.Circle((0, 0.5), 2, color='pink',alpha=0.5, fill=True);
plt.plot([0],[1.8],marker='*', color='blue',linewidth=8,markersize=12);
plt.plot(np.array(avg).mean(axis=0)[:,0],np.array(avg).mean(axis=0)[:,1],marker='o', color='green',linewidth=8,markersize=12);
a =  plt.gca()
a.add_patch(c)
xax = a.axes.get_yaxis()
xax = xax.set_visible(False)
plt.axis('off')
plt.rc('axes.spines', **{'bottom':True, 'left':True, 'right':False, 'top':False})
plt.savefig("distribution_2.png");
```
